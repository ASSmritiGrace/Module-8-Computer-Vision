```python
# Gaussian Blur
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('image 4.jpg')
```

```python
# GaussianBlur
blurred = cv2.GaussianBlur(image,(5,5),0)
```

```python
# Median Blur 

median_blurred = cv2.medianBlur(image,5)
```

```python
# Bilateral Filtering

filtered =cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
# Sharpening Filters
mask=cv2.subtract(image,blurred)
```

```python
sharpened=cv2.addWeighted(image, 1.0,mask,1.5,0)
```

```python
plt.subplot(1,5,1)
plt.title('Original')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,5,2)
plt.title('Gaussia Blurred')
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1,5,3)
plt.title('Median Blurred')
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,,4)
plt.title('Bilateral Filtering ')
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1,4,4)
plt.title('Sharpened ')
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
