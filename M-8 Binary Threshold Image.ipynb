```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images3.jpg')
```

```python
xyz=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurred_image = cv2.GaussianBlur(xyz, (5,5), 0)
```

```python
TH=100
max_TH=255
_, Threshold=cv2.threshold(blurred_image,TH,max_TH,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.title("Original Image:")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("B&W Image")
plt.imshow(xyz,cmap='gray')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.title("Binary Image (Thresholding)")
plt.imshow(Threshold / 255.0,cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
